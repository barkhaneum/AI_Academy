페이스북의 FastText

FastText는 단어를 표현과 문장 분류의 효율적인 학습을 위한 라이브러리 입니다. FastText는 페이스북이 2016년에 발표한 것으로 원래 단어를 부분단어(subword)의 벡터들로 표현한다는 점을 제외하고는 Word2Vec과 거의 유사합니다. 노이즈가 많은 말뭉치에 강점을 지니고, 학습시간에 있어서 빠르다는 장점을 가집니다.

FastText같은 경우 카카오톡 플러스 친구를통해서 FastText로 챗봇 기능을 구현하고
